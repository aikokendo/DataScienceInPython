```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
# Load the raw data
df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
```

```python
# Create a new_df with target columns
df['Date'] = pd.to_datetime(df['Date'])
df['MM-DD'] = df['Date'].dt.strftime('%m-%d')
df['Data_Value'] = df['Data_Value']/10

filt_2014 = ((df['Date'].dt.year >= 2005) & (df['Date'].dt.year <= 2014))
filt_2015 = (df['Date'].dt.year == 2015)
filt_Feb = ((df['Date'].dt.month == 2) & (df['Date'].dt.day == 29))

# select year 2005-2014
grp_2014 = df[filt_2014 & ~filt_Feb].groupby(['MM-DD'])
df_2014 = grp_2014['Data_Value'].agg(['max','min'])

# select year 2015
grp_2015 = df[filt_2015].groupby(['MM-DD'])
df_2015 = grp_2015['Data_Value'].agg(['max','min'])
```

```python
new_df = pd.merge(df_2014, df_2015, how = 'inner', left_index = True, right_index = True)
new_df.rename(columns={'max_x':'max_2014', 'min_x':'min_2014', 'max_y':'max_2015', 'min_y':'min_2015'}, inplace = True)
new_df['Date'] = list(map(lambda date: '2015-'+date, new_df.index.tolist()))
new_df['Date'] = pd.to_datetime(new_df['Date'])


```

```python
import numpy as np

def high_record():
    lstdate = list()
    lstvalue = list()
    for date in new_df['Date']:
        index = new_df.index[new_df['Date'] == date]
        max_2015 = new_df.loc[index, 'max_2015'].values
        max_2014 = new_df.loc[index, 'max_2014'].values
        if max_2015 - max_2014>0:
            lstdate.append(date)
            lstvalue.append(max_2015[0])
    zip_generator_h = zip(lstdate,lstvalue)
    return zip_generator_h

def low_record():
    lowlist = dict()
    for date in new_df['Date']:
        index = new_df.index[new_df['Date'] == date]
        min_2015 = new_df.loc[index, 'min_2015'].values
        min_2014 = new_df.loc[index, 'min_2014'].values
        if min_2014 - min_2015>0:
            lowlist[date] = min_2015[0]
    lowdata = list(lowlist.items())
    lowarray = np.array(lowdata)
    return lowarray

```

```python
%matplotlib notebook
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
   
x_2014 = new_df['Date'].tolist()
y_2014max = new_df['max_2014'].tolist()
y_2014min = new_df['min_2014'].tolist()

plt.figure()
plt.plot(x_2014, y_2014max , '-', label = '2005-2014 Highest Record')
plt.plot(x_2014, y_2014min, '-', label = '2005-2014 Lowest Record' )
plt.gca().fill_between(x_2014, y_2014max, y_2014min, facecolor = 'gray', alpha = 0.2)

xhigh, yhigh = zip(*high_record())
plt.plot(xhigh,yhigh,'o', color = 'tomato', label = '2015 New Record for High Temp')

xlow = low_record()[:,0]
ylow = low_record()[:,1]
plt.plot(xlow,ylow,'o', color = 'dodgerblue',label = '2015 New Record for Low Temp')

# Convert x-axis to Month
locator = mdates.MonthLocator()  # every month
fmt = mdates.DateFormatter('%b')
X = plt.gca().xaxis
X.set_major_locator(locator)
X.set_major_formatter(fmt)

# Add label
plt.ylabel('Temperature in $Â°C$')
plt.title('Temperature Record for Ann Arbor, Michigan Area')
plt.legend(loc ='lower right')
plt.ylim(-50, 45)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.tick_params(top = 'off', bottom = 'off', left = 'off', right = 'off')

plt.savefig('myfig')
```
